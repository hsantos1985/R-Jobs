##SCRIPT NO R CONTENDO CONCEITOS TESTE DE AUTOCORRELAÇÃO EM SÉRIES TEMPORAIS

```r
#CRIANDO A SÉRIE COM DADOS ALEATÓRIOS
set.seed(8)
dados = rnorm(72)
serie = ts(dados,start = c(2015,1), end=c(2020,12), frequency=12)
print(serie)
```

```r
#PLOTANDO A SERIE

plot(serie)
```

```r
#PLOTANDO OA DIAGRAMAS DE AUTOCORRELAÇÃO - ACF E AUTOCORRELAÇÃO PARCIAL - PACF
```

```r
#ACF
acf(serie)
```

```r
#pACF
pacf(serie)
```

```r
# Teste de Autocorrelação (Ljung-Box)

# Ho = não autocorrelacionado: p > 0.05
# Ha = é autocorrelacionado: p <= 0.05
Box.test (serie, type = "Ljung")
```

```r
#como p-value é maior que 0,05 a série não tem autocorrelação
```

```r
# Estacionaridade
# Teste pp (Phillips-Perron)

# Ho = é estacionário: p > 0.05
# Ha = não é estacionério: p <= 0.05
```

```r
install.packages('urca')
```

```r
library("urca")
pp <- ur.pp(manchas_solares)
summary(pp)
```

```r
#a série não é estacionária
```

```r
#autocorrelação
acf(manchas_solares)
```

```r
pacf(manchas_solares)
```

```r
#os dados possuem autocorrelção, pel aanálise visual
```
