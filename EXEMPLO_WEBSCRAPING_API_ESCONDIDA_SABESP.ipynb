```r
#PACOTES
install.packages("janitor")
```

```r
#BIBLIOTECAS
library(dplyr)
library(httr)
library(jsonlite)
library(janitor)
```

```r
#CRIANDO A REQUISIÇÃO GET####
#parâmetros####
dia<-
  '2024-07-17'
u_base<-
  "https://mananciais.sabesp.com.br/api/Mananciais/ResumoSistemas/"
u_dia<-
  paste0(u_base,dia)
r_dia<-
  GET(u_dia)
l_dia<-
  content(r_dia)
l_dia$ReturnObj$sistemas
```

```r
#transformando em dataframe
l_dia<-
  content(r_dia, simplifyDataFrame=TRUE)
l_dia$ReturnObj$sistemas %>%
  tibble::tibble()
```

```r
#acessando os elementso da lista com o pluck()
d_dia<-
l_dia %>%
  purrr::pluck("ReturnObj", "sistemas") %>%
  tibble::tibble() %>%
  janitor::clean_names()

d_dia
```

```r
#CRIANDO UMA FUNÇÃO PARA AUTOMATIAR O PROCESSO####
sabesp_baixar_dia<-
  function(dia){
    u_base<-"https://mananciais.sabesp.com.br/api/Mananciais/ResumoSistemas/"
    u_dia<-paste0(u_base,dia)
    r_dia<-GET(u_dia)
    l_dia<-content(r_dia)
    #transformando em dataframe
    l_dia<-content(r_dia, simplifyDataFrame=TRUE)
    d_dia<-l_dia %>%
      purrr::pluck("ReturnObj", "sistemas") %>%
      tibble::tibble() %>%
      janitor::clean_names()
    d_dia

  }
#testado a função
sabesp_baixar_dia("2025-06-03")
```
