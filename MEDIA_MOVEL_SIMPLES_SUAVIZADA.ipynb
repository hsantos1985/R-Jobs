##SCRIPT NO R PARA SUAVIZAÇÃO POR MÉDIA MÓVEL SIMPLES E SUAVIZADA

#####MÉDIA MÓVEL COM SUAVIZAÇÃO

```r
# Desvios da Temperatura Global
install.packages('astsa')
library(astsa)
```

```r
#configuração dos gráficos
options(repr.plot.width = 19, repr.plot.height = 10)
```

```r
gtemp_both
```

```r
temp_global<-
  ts(gtemp_both, start = c(1850), end = c(2023), frequency = 1)
```

```r
print(temp_global)
```

```r
plot(temp_global, type="l", ylab="Desvios da Temperatura",col="blue")
```

```r
#criando a média móvel
install.packages('forecast')
library(forecast)
```

```r
#fazendo a média móvel de 7 períodos
temp_global2<-
  ma(temp_global, order = 7, centre = TRUE)
plot(temp_global2)
```

```r
temp_global3<-
  ma(temp_global, order = 20, centre = FALSE)
plot(temp_global3)
```

```r
temp_global31<-
  ma(temp_global, order = 20, centre = FALSE)
plot(temp_global31)
```

```r
temp_global4<-
  ma(temp_global, order = 51, centre = TRUE)
plot(temp_global4)
```

#####COMPARANDO AS MÉDIAS MÓVEIS
