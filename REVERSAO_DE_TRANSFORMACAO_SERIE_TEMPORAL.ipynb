##SCRIPT NO R CONTENDO CONCEITOS DE REVERSÃO DE TRANSFORMAÇÃO E DIFERENCIAÇÃO DE SÉRIES TEMPORAIS

####1-TESTANDO NORMALIDADE E REALIZANDO TRANSFORMAÇÕES

```r
#configuração dos gráficos
options(repr.plot.width = 19, repr.plot.height = 10)
```

```r
# Nmúero de passageiros aéreos entre 1949 a 1960

passageiros <- AirPassengers
passageiros
```

```r
#transformação da série em log natural
passageiros2<-
  log(passageiros)

passageiros2
```

```r
#reversão da transformação
reversao_passageiros2<-
  exp(passageiros2)
reversao_passageiros2
```

```r
#transformação com logaritmo base 10
passageiros3<-
  log10(passageiros)

passageiros3
```

```r
#reversao da transformação
reversao_passageiros3<-
  10^(passageiros3)

reversao_passageiros3
```

```r
#transformação raiz cúbica
passageiros4<-
  passageiros^(1/3)

passageiros4
```

```r
#reversão raiz cúbica
reversao_passageiros4<-
  passageiros4^(3)
reversao_passageiros4

```

####DIFERENCIAÇÃO


```r
#fazendo a diferenciação
dif_passageiros<-
  diff(passageiros)

dif_passageiros
```

```r
#revertendo a diferenciação
reversao_dif_passageiros<-
  diffinv(dif_passageiros, xi = 112)
reversao_dif_passageiros
```

```r
#FAZENDO UMA SEGUNDA DIFERENCIAÇÃO
segunda_dif_passageiros<-
  diff(dif_passageiros)
segunda_dif_passageiros
```

```r
#reversao da seguna diferenciação
reversao_segunda_dif_passageiros1<-
    diffinv(segunda_dif_passageiros, xi =6)

reversao_segunda_dif_passageiros1
```

```r
reversao_segunda_dif_passageiros2<-
    diffinv(reversao_segunda_dif_passageiros1, xi =112)

reversao_segunda_dif_passageiros2
```
