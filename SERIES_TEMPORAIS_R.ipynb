##SCRIPT NO R CONTENDO CONCEITOS SOBRE ESTACIONARIDADE DE SERIES

```r
#serie anual####
set.seed(10)
dados1<-
  rnorm(41)
```

```r
serie1<-
  ts(dados1,start = c(1980), end = c(2020), frequency = 1)

print(serie1)
plot(serie1)
```

```r
#NORMALIDADE
qqnorm(serie1)
qqline(serie1)
```

### Teste KPSS (Kwiatkowski-Phillips-Schmidt-Shin)

```r
install.packages('urca')
```

```r
# Ha = não é estacionária: teste estatístico > valor crítico
# Ho = é estacionária:  teste estatístico <= valor crítico
library('urca')

kpss<-
  ur.kpss(serie1)

summary(kpss)
```

```r
#série estacionária
```

####Teste pp (Philips-Perron)

```r
# Ho = é estacionária: p > 0.05
# Ha = não é estacionária: p <= 0.05

pp<-
  ur.pp(serie1)
summary(pp)
```

```r
#serie estacionária
```

####Teste df (Dickey Fuller)

```r
# Ho = não é estacionária: teste estatístico > valor crítico
# Ha = é estacionária:  teste estatístico <= valor crítico

df<-
  ur.df(serie1)
summary(df)
```

```r
#serie estacionária
```
