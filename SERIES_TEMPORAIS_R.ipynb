##SCRIPT NO R CONTENDO CONCEITOS SOBRE ESTACIONARIDADE DE SERIES

```r
#serie anual####
set.seed(10)
dados1<-
  rnorm(41)
```

```r
serie1<-
  ts(dados1,start = c(1980), end = c(2020), frequency = 1)

print(serie1)
plot(serie1)
```

```r
#NORMALIDADE
qqnorm(serie1)
qqline(serie1)
```

### Teste KPSS (Kwiatkowski-Phillips-Schmidt-Shin)

```r
install.packages('urca')
```

```r
# Ha = no  estacionria: teste estatstico > valor crtico
# Ho =  estacionria:  teste estatstico <= valor crtico
library('urca')

kpss<-
  ur.kpss(serie1)

summary(kpss)
```

```r
#srie estacionria
```

####Teste pp (Philips-Perron)

```r
# Ho =  estacionria: p > 0.05
# Ha = no  estacionria: p <= 0.05

pp<-
  ur.pp(serie1)
summary(pp)
```

```r
#serie estacionria
```

####Teste df (Dickey Fuller)

```r
# Ho = no  estacionria: teste estatstico > valor crtico
# Ha =  estacionria:  teste estatstico <= valor crtico

df<-
  ur.df(serie1)
summary(df)
```

```r
#serie estacionria
```
