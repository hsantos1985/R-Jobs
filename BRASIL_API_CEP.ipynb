#OBTENÇÃO DE CEP DA API - "BRASIL API"

####PARÂMETROS

```r
#BIBLIOTECAS
library(dplyr)
library(httr)
library(jsonlite)
```

```r
u_base<-
  "https://brasilapi.com.br/api"
```

####CEP

```r
endpoint_cep<-
  "/cep/v2/"

cep<-
  "01310930"

u_cep<-
  paste0(u_base,endpoint_cep,cep)

r_cep<-
  GET(u_cep)
```

```r
#obtenção dos resultados

content(r_cep)

content(r_cep, as="text")

content(r_cep, as="raw")

content(r_cep, as="parsed")

content(r_cep, as="text") %>%
  fromJSON()
```

####TABELA FIPE

```r
endpoint_fipe<-
  '/fipe/marcas/v1/'

tipo_veiculos<-
  'carros'

u_fipe<-
  paste0(u_base,endpoint_fipe,tipo_veiculos)
```

```r
r_fipe<-
  GET(u_fipe, write_disk("C:\\Users\\SSD480GB\\OneDrive\\Documentos\\03_EDUCAIONAL\\HOTMART\\01_CURSO_WEBSCRAPING\\DADOS\\fipe_marcas.json", overwrite = TRUE))
```

```r
content(r_fipe, as='text') %>%
  fromJSON(simplifyDataFrame = TRUE)


content(r_fipe,simplifyDataFrame = TRUE)


content(r_fipe,simplifyDataFrame = TRUE) %>%
  tibble::as_tibble()
```
