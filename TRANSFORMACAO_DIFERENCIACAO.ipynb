####1-TESTANDO NORMALIDADE E REALIZANDO TRANSFORMAÇÃO

```r
#configuração dos gráficos
options(repr.plot.width = 19, repr.plot.height = 10)
```

```r
#criação de série mensal
#dados normalmente distribuidos
set.seed(6)
dados <-
  rnorm(72, 10, 1)
```

```r
serie<-
  ts(dados, start=c(2015,1), end=c(2020,12), frequency=12)
```

```r
print(serie)
```

```r
plot(serie)
```

```r
summary(serie)
```

```r
#NORMALIDADE
qqnorm(serie)
qqline(serie)
```

```r
#APLICANDO SHAPIRO
#hipóteses
# Ho = distribuição normal : p > 0.05
# Ha = distribuição != normal : p <= 0.05

shapiro.test(serie)
```

```r
#distribuição não é normal
```

```r
hist(serie)
```

####UTILIZANDO O DF CO2

```r
install.packages("repr")
library(repr)
```

```r
#UTILIZANDO DF DO PRÓPRIO R
help("co2")
```

```r
co2
```

```r
concentracao<-
  co2
```

```r
print(concentracao)
```

```r
#configuração dos gráficos
options(repr.plot.width = 12, repr.plot.height = 13)
```

```r
plot(concentracao)
```

```r
#REALIZANDO A DECOMPOSIÇÃO DO DF
install.packages('forecast')
```

```r
library(forecast)
```

```r
#decomposicao aditiva
decomposicao<-
  decompose(concentracao)
```

```r
plot(decomposicao)
```

```r
#decomposicao multiplicativa
decomposicao_mult<-
  decompose(concentracao, type = 'mult')
```

```r
plot(decomposicao_mult)
```

####DF TEMPERATURA GLOBAL

```r
install.packages('astsa')
library(astsa)
```

```r
help("gtemp_both")
```

```r
temp_global<-
ts(gtemp_both, start = 1, end = 87, frequency = 2)
```

```r
print(temp_global)
```

```r
plot(temp_global, type="l", ylab="Desvios da Temperatura",col="blue")
```

```r
#decomposicao da serie

decomposicao_temperatura<-
  decompose(temp_global, type = 'mult')
```

```r
plot(decomposicao_temperatura)
```

```r
autoplot(decomposicao_temperatura)
```
