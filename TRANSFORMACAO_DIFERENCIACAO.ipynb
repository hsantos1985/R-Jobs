####1-TESTANDO NORMALIDADE E REALIZANDO TRANSFORMAÇÕES

```r
#configuração dos gráficos
options(repr.plot.width = 19, repr.plot.height = 10)
```

```r
#criação de série mensal
#dados normalmente distribuidos
set.seed(6)
dados <-
  rnorm(72, 10, 1)
```

```r
serie<-
  ts(dados, start=c(2015,1), end=c(2020,12), frequency=12)
```

```r
print(serie)
```

```r
plot(serie)
```

```r
summary(serie)
```

```r
#NORMALIDADE
qqnorm(serie)
qqline(serie)
```

```r
#APLICANDO SHAPIRO
#hipóteses
# Ho = distribuição normal : p > 0.05
# Ha = distribuição != normal : p <= 0.05

shapiro.test(serie)
```

```r
#distribuição não é normal
```

```r
hist(serie)
```

```r
#REALIZANDO TRANSFORMAÇAO NA SERIE
serie2<-
  log(serie)
```

```r
print(serie2)
```

```r
plot(serie2)
```

```r
summary(serie2)
```

```r
hist(serie2)
```

```r
#NORMALIDADE
qqnorm(serie)
qqline(serie)
```

```r
#APLICANDO SHAPIRO
#hipóteses
# Ho = distribuição normal : p > 0.05
# Ha = distribuição != normal : p <= 0.05

shapiro.test(serie2)
```

```r
#DISTRIBUIÇÃO É NORMAL AGORA
```

```r
#criando a serie 3
```

```r
#EXEMPLO DE ARTIFÍCIO PARA NÃO DEIXAR NaN nas séries com valores negativos (transfofração para valores com raiz par)
sign(-27)*abs(-27^(1/3))
```

```r
#UTILIZANDO A TRANSFORMAÇÃO COM RAÍZ CÚBICA (PARA QUANDO OS DADOS SÃO NEGATIVOS E IGUAIS A 0)
serie3<-
  #utilizando a transformmação só como aplicação
  sign(serie)*abs(serie)^(1/3)
```

```r
print(serie3)
```

```r
plot(serie3)
```

```r
hist(serie3)
```

```r
summary(serie3)
```

```r
#APLICANDO SHAPIRO
#hipóteses
# Ho = distribuição normal : p > 0.05
# Ha = distribuição != normal : p <= 0.05

shapiro.test(serie3)
```

```r
#a série é normal
```

```r
#FAZENDO O  TESTE DE ESTACIONARIDADE
# Teste pp (Philips-Perron)

# Ho = é estacionária: p > 0.05
# Ha = não é estacionária: p <= 0.05
library(urca)
pp<-
  ur.pp(serie2)
summary(pp)
```

```r
#como p-value: 0.2776 e maior que 0,05 a série é estacionária
```
